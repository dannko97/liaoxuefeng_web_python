## 用户定义的可调用类型

```python
import random
class BingoCage():
    def __init__(self, items):
        self._items = list(items)
        random.shuffle(self._items)
        
    def pick(self):
        # try:
            return self._items.pop()
#         except IndexError:
#             rasie LookupError('pick from empty BingoCage')
    
    def __call__(self):
        return self.pick()
```

```python
bg = BingoCage(range(3))
```

```python
bg.pick()
```

```python
bg()
```
