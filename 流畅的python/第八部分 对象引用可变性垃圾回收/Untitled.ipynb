构造方法或 [:] 做的是浅复制（即复制了最外层容器，副本中 的元素是源容器中元素的引用）。如果所有元素都是不可变的，那么这 样没有问题，还能节省内存。但是，如果有可变的元素，可能就会导致 意想不到的问题。

对元组来说，+= 运算符创建一个新元组，然后重新绑定给变量 l2[2]。这等同于 l2[2] = l2[2] + (10, 11)。现在，l1 和 l2 中最 后位置上的元组不是同一个对象。

```python
def f(a, b):
    a += b 
    return a
x,y = 1, 2
f(x,y)
a,b = [1,2],[3,4]
f(a,b)
t,u=(10,20),(30,40)
f(t,u)
```
